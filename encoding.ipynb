```python
f = open("broken.txt", 'rb')
encoding_list = ['euc_kr','utf-8', 'utf-16', 'utf-32']
b = 0
c = 0
```

```python
while True:
    a = f.readline()
    if not a: break
    for encoding in encoding_list:
        try:
            if a[0] == 0:
                print(a[1:-1].decode(encoding))
            else:
                print(a[:-1].decode(encoding))
        except UnicodeDecodeError:
            pass
        else:
            break
            
f.close()
```
